```python
import math
import copy

X, O, EMPTY = "X", "O", None

def initial_state(n):
    return [[EMPTY for _ in range(n)] for _ in range(n)]

def player(board):
    x_count = sum(row.count(X) for row in board)
    o_count = sum(row.count(O) for row in board)
    return X if x_count == o_count else O

def actions(board):
    n = len(board)
    return [(r, c) for r in range(n) for c in range(n) if board[r][c] == EMPTY]

def result(board, action):
    new_board = copy.deepcopy(board)
    new_board[action[0]][action[1]] = player(board)
    return new_board

def check_winner(board):
    n = len(board)
    for i in range(n):
        if all(board[i][j] == board[i][0] != EMPTY for j in range(n)): return board[i][0]
        if all(board[j][i] == board[0][i] != EMPTY for j in range(n)): return board[0][i]
    if all(board[i][i] == board[0][0] != EMPTY for i in range(n)): return board[0][0]
    if all(board[i][n-1-i] == board[0][n-1] != EMPTY for i in range(n)): return board[0][n-1]
    return None

def terminal(board):
    return check_winner(board) is not None or not actions(board)

def evaluation(board):
    res = check_winner(board)
    if res == X: return 100
    if res == O: return -100
    return 0

def minimax(board, depth, is_maximizing, max_depth):
    if terminal(board) or depth == max_depth:
        return evaluation(board)

    if is_maximizing:
        return max(minimax(result(board, a), depth + 1, False, max_depth) for a in actions(board))
    else:
        return min(minimax(result(board, a), depth + 1, True, max_depth) for a in actions(board))

def main():
    n = int(input("Nhập kích thước N cho Minimax: "))
    board = initial_state(n)
    # Giới hạn độ sâu để tránh treo máy (N càng lớn depth càng nhỏ)
    depth_limit = 3 if n <= 3 else 2 

    while not terminal(board):
        for r in board: print(r)
        if player(board) == X:
            r, c = map(int, input("Lượt X (hàng cột): ").split())
            board[r][c] = X
        else:
            print("AI đang tính toán...")
            move = max(actions(board), key=lambda a: minimax(result(board, a), 1, False, depth_limit))
            board[move[0]][move[1]] = O
    print("Kết thúc:", check_winner(board) if check_winner(board) else "Hòa")

if __name__ == "__main__":
    main()
```

```python
import math
import copy

X, O, EMPTY = "X", "O", None

def initial_state(n):
    return [[EMPTY for _ in range(n)] for _ in range(n)]

def player(board):
    x_count = sum(row.count(X) for row in board)
    o_count = sum(row.count(O) for row in board)
    return X if x_count == o_count else O

def actions(board):
    n = len(board)
    return [(r, c) for r in range(n) for c in range(n) if board[r][c] == EMPTY]

def result(board, action):
    new_board = copy.deepcopy(board)
    new_board[action[0]][action[1]] = player(board)
    return new_board

def check_winner(board):
    n = len(board)
    for i in range(n):
        if all(board[i][j] == board[i][0] != EMPTY for j in range(n)): return board[i][0]
        if all(board[j][i] == board[0][i] != EMPTY for j in range(n)): return board[0][i]
    if all(board[i][i] == board[0][0] != EMPTY for i in range(n)): return board[0][0]
    if all(board[i][n-1-i] == board[0][n-1] != EMPTY for i in range(n)): return board[0][n-1]
    return None

def terminal(board):
    return check_winner(board) is not None or not actions(board)

def alpha_beta(board, depth, alpha, beta, is_maximizing, max_depth):
    if terminal(board) or depth == max_depth:
        res = check_winner(board)
        return 100 if res == X else -100 if res == O else 0

    if is_maximizing:
        v = -math.inf
        for a in actions(board):
            v = max(v, alpha_beta(result(board, a), depth + 1, alpha, beta, False, max_depth))
            alpha = max(alpha, v)
            if beta <= alpha: break
        return v
    else:
        v = math.inf
        for a in actions(board):
            v = min(v, alpha_beta(result(board, a), depth + 1, alpha, beta, True, max_depth))
            beta = min(beta, v)
            if beta <= alpha: break
        return v

def play_alpha_beta():
    n = int(input("Nhập kích thước N cho Alpha-Beta: "))
    board = initial_state(n)
    depth_limit = 4 if n <= 3 else 3

    while not terminal(board):
        for r in board: print(r)
        curr = player(board)
        if curr == X:
            r, c = map(int, input(f"Lượt {X} (hàng cột): ").split())
            board[r][c] = X
        else:
            print("AI (Alpha-Beta) đang tính toán...")
            best_val = math.inf # Giả sử AI là O (MIN)
            best_move = None
            for a in actions(board):
                val = alpha_beta(result(board, a), 1, -math.inf, math.inf, True, depth_limit)
                if val < best_val:
                    best_val = val
                    best_move = a
            board[best_move[0]][best_move[1]] = O
    print("Kết thúc:", check_winner(board) if check_winner(board) else "Hòa")

if __name__ == "__main__":
    play_alpha_beta()
```
