```python
import copy
import math
import numpy

X, O, EMPTY = "X", "O", None

def initial_state():
    return [[EMPTY] * 3 for _ in range(3)]

def player(board):
    count = sum(row.count(X) + row.count(O) for row in board)
    return X if count % 2 == 0 else O

def actions(board):
    return {(i, j) for i in range(3) for j in range(3) if board[i][j] == EMPTY}

def result(board, action):
    new_board = copy.deepcopy(board)
    new_board[action[0]][action[1]] = player(board)
    return new_board

def winner(board):
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != EMPTY: return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != EMPTY: return board[0][i]
    if board[0][0] == board[1][1] == board[2][2] != EMPTY: return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != EMPTY: return board[0][2]
    return None

def terminal(board):
    return winner(board) is not None or not actions(board)

def utility(board):
    res = winner(board)
    return 1 if res == X else -1 if res == O else 0

def maxValue(state):
    if terminal(state): return utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action)))
    return v

def minValue(state):
    if terminal(state): return utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action)))
    return v

def minimax(board):
    curr = player(board)
    if curr == X:
        return max(actions(board), key=lambda a: minValue(result(board, a)))
    return min(actions(board), key=lambda a: maxValue(result(board, a)))

# --- PHẦN CHẠY TRÒ CHƠI ---
board = initial_state()
print("CHÀO MỪNG BẠN ĐẾN VỚI TICTACTOE")
user_input = input("Chọn quân (X đi trước, O đi sau): ").upper()
user = X if user_input == "X" else O
ai = O if user == X else X

while not terminal(board):
    curr_p = player(board)
    if curr_p == ai:
        print("\nAI đang tính toán...")
        move = minimax(board)
        board = result(board, move)
        print(numpy.array(board))
    else:
        print(f"\nLượt của bạn ({user})")
        valid = False
        while not valid:
            try:
                # Dùng thông báo rõ ràng để Notebook hiện ô nhập
                r = int(input("Nhập Hàng (0, 1, hoặc 2): "))
                c = int(input("Nhập Cột (0, 1, hoặc 2): "))
                if board[r][c] == EMPTY:
                    board = result(board, (r, c))
                    print(numpy.array(board))
                    valid = True
                else:
                    print("Ô đã có quân, mời nhập lại!")
            except:
                print("Lỗi! Vui lòng nhập số 0, 1 hoặc 2.")

res = winner(board)
print("\nKẾT QUẢ:", f"{res} thắng!" if res else "Hòa cờ!")
```

```python
import os
import math

def GetWinner(board):
    # Các tổ hợp thắng (indices)
    win_combinations = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8], 
        [0, 3, 6], [1, 4, 7], [2, 5, 8], 
        [0, 4, 8], [2, 4, 6]             
    ]
    for combo in win_combinations:
        if board[combo[0]] == board[combo[1]] == board[combo[2]]:
            return board[combo[0]]
    return None

def PrintBoard(board):
    os.system('cls' if os.name == 'nt' else 'clear')
    print(f"\n    {board[0]} | {board[1]} | {board[2]}")
    print("    --+---+--")
    print(f"    {board[3]} | {board[4]} | {board[5]}")
    print("    --+---+--")
    print(f"    {board[6]} | {board[7]} | {board[8]}\n")

def GetAvailableCells(board):
    
    return [cell for cell in board if cell != "X" and cell != "O"]

def minimax(position, depth, alpha, beta, isMaximizing):
    winner = GetWinner(position)
    if winner == "X":
        return 10 - depth
    if winner == "O":
        return -10 + depth
    if len(GetAvailableCells(position)) == 0:
        return 0

    if isMaximizing:
        maxEval = -math.inf
        for cell in GetAvailableCells(position):
            idx = cell - 1
            position[idx] = "X"
            eval = minimax(position, depth + 1, alpha, beta, False)
            position[idx] = cell
            maxEval = max(maxEval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                break
        return maxEval
    else:
        minEval = math.inf
        for cell in GetAvailableCells(position):
            idx = cell - 1
            position[idx] = "O"
            eval = minimax(position, depth + 1, alpha, beta, True)
            position[idx] = cell
            minEval = min(minEval, eval)
            beta = min(beta, eval)
            if beta <= alpha:
                break
        return minEval

def FindBestMove(currentPosition, AI_Player):
    bestMove = -1
    if AI_Player == "X":
        bestVal = -math.inf
        for cell in GetAvailableCells(currentPosition):
            idx = cell - 1
            currentPosition[idx] = "X"
            moveVal = minimax(currentPosition, 0, -math.inf, math.inf, False)
            currentPosition[idx] = cell
            if moveVal > bestVal:
                bestVal = moveVal
                bestMove = cell
    else:
        bestVal = math.inf
        for cell in GetAvailableCells(currentPosition):
            idx = cell - 1
            currentPosition[idx] = "O"
            moveVal = minimax(currentPosition, 0, -math.inf, math.inf, True)
            currentPosition[idx] = cell
            if moveVal < bestVal:
                bestVal = moveVal
                bestMove = cell
    return bestMove

def main():
    print("--- CHÀO MỪNG ĐẾN VỚI TICTACTOE ALPHA-BETA ---")
    while True:
        player_choice = input("Bạn muốn chơi X hay O? (X đi trước): ").strip().upper()
        if player_choice in ["X", "O"]:
            break
    
    AI = "O" if player_choice == "X" else "X"
    currentGame = [*range(1, 10)]
    currentTurn = "X" # X luôn đi trước
    
    while True:
        PrintBoard(currentGame)
        
        # Kiểm tra kết thúc game
        winner = GetWinner(currentGame)
        if winner:
            print(f"KẾT THÚC: {winner} CHIẾN THẮNG!!!")
            break
        if len(GetAvailableCells(currentGame)) == 0:
            print("KẾT THÚC: HÒA CỜ!")
            break

        if currentTurn == AI:
            print(f"AI ({AI}) đang tính toán...")
            if len(GetAvailableCells(currentGame)) == 9:
                cell = 1
            else:
                cell = FindBestMove(currentGame, AI)
            
            currentGame[cell - 1] = AI
            currentTurn = player_choice
        else:
            while True:
                try:
                    humanInput = int(input(f"Lượt của bạn ({player_choice}) - Nhập số (1-9): "))
                    if humanInput in currentGame:
                        currentGame[humanInput - 1] = player_choice
                        currentTurn = AI
                        break
                    else:
                        print("Ô không hợp lệ hoặc đã có người đi!")
                except ValueError:
                    print("Vui lòng chỉ nhập số từ 1 đến 9!")

if __name__ == "__main__":
    main()
```
