```python
import math
import string
def nhapmang1():
    a = []
    n = int(input("Nhap so luong mang muon khoi tao:"))
    for i in  range(n):
        giatri = float(input(f"Nhap gia tri thu {i} la:"))
        a.append(giatri)
    return a

def nhapmang2():
    b = []
    m = int(input("Nhap so luong mang muon khoi tao:"))
    for j in range(m):
        giatri = float(input(f"Nhap gia tri thu {j} la:"))
        b.append(giatri)
    return b
def tronmang(a, b):
    mang_ket_qua = []
    min_len = min(len(a), len(b))
    for i in range(min_len):
        gia_tri_min = min(a[i], b[i])
        mang_ket_qua.append(gia_tri_min)
    if len(a) > len(b):
        mang_ket_qua.extend(a[min_len:])
    else:
        mang_ket_qua.extend(b[min_len:])
    return mang_ket_qua

    return mang_ket_qua
def main():
    mang_a = nhapmang1()
    mang_b = nhapmang2()
    ketqua = tronmang(mang_a, mang_b)
    print(f"Cac phan tu co trong mang la: {ketqua}")

if __name__ == "__main__":
    main()
```

```python
def Tao_Mang():
    a = []
    rows = int(input("Nhap so luong hang muon tao: "))
    colums = int(input("Nhap so luong cot muon khoi tao: "))
    for i in range(rows):
        row = []
        for j in range(colums):
            value = int(input(f"Nhap gia tri thu {i}{j} la: "))
            row.append(value)
        a.append(row)
    return a

def tong_tam_giac_tren_duong_cheo_phu(a):
    n = len(a) 
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    return tong

def tri_tuyet_doi_ma_tran(a):
    # Sao chép để không làm hỏng mảng gốc
    res = [row[:] for row in a]
    for i in range(len(res)):
        for j in range(len(res[0])):
            res[i][j] = abs(res[i][j])
    return res

def thay_chan_bang_x(a, x):
    res = [row[:] for row in a]
    for i in range(len(res)):
        for j in range(len(res[0])):
            if res[i][j] % 2 == 0:
                res[i][j] = x
    return res

def kiem_tra_toan_chan(a):
    for row in a:
        for item in row:
            if item % 2 != 0:
                return False 
    return True

def kiem_tra_doi_xung(a):
    n = len(a)
    for i in range(n):
        for j in range(i + 1, n):
            if a[i][j] != a[j][i]:
                return False
    return True

def duong_cheo_chinh_tang_dan(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][i] >= a[i+1][i+1]:
            return False
    return True

def xuat_tam_giac_duoi_phu(a):
    n = len(a)
    print("Tam giac duoi duong cheo phu:")
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(a[i][j], end=" ")
            else:
                print(" ", end=" ")
        print()

def duong_cheo_phu_giam_dan(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][n - 1 - i] <= a[i+1][n - 2 - i]:
            return False
    return True

def main():
    # 1. Khoi tao
    mang2chieu = Tao_Mang()
    
    # a. Tong tam giac tren duong cheo phu
    print(f"\nTong cac phan tu thuoc tam giac tren duong cheo phu: {tong_tam_giac_tren_duong_cheo_phu(mang2chieu)}")
    
    # b. Tri tuyet doi
    print(f"Ma tran sau khi lay tri tuyet doi: {tri_tuyet_doi_ma_tran(mang2chieu)}")
    
    # c. Thay chan bang x
    x_val = float(input("Nhap gia tri x muon thay the: "))
    print(f"Ma tran sau khi thay phan tu chan bang x: {thay_chan_bang_x(mang2chieu, x_val)}")
    
    # d. Kiem tra toan chan
    print(f"Kiem tra ma tran co toan so chan: {kiem_tra_toan_chan(mang2chieu)}")
    
    # e. Kiem tra doi xung
    print(f"Kiem tra ma tran co doi xung: {kiem_tra_doi_xung(mang2chieu)}")
    
    # f. Duong cheo chinh tang dan
    print(f"Kiem tra duong cheo chinh tang dan: {duong_cheo_chinh_tang_dan(mang2chieu)}")
    
    # g. Xuat tam giac duoi phu
    xuat_tam_giac_duoi_phu(mang2chieu)
    
    # h. Duong cheo phu giam dan
    print(f"Kiem tra duong cheo phu giam dan: {duong_cheo_phu_giam_dan(mang2chieu)}")

if __name__ == "__main__":
    main()
```
